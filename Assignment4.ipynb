```python
class Person:
    def __init__(self, name, age, country):
        self.name = name
        self.age = age
        self.country = country

    def pdetails(self):
        print("Person Details:")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Country: {self.country}")
        
p1 = Person("Ahsan", 25, "Pakistan")
p2 = Person("Tahir", 30, "Canada")

p1.display_details()
p2.display_details()


class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)


r1 = Rectangle(10, 5)

print("Width:", r1.width)
print("Height:", r1.height)
print("Area:", r1.area())
print("Perimeter:", r1.perimeter())



class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def vdetails(self):
        print(f"Make: {self.make}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")

class Car(Vehicle):
    def __init__(self, make, model, year, doors):
        super().__init__(make, model, year)
        self.doors = doors

    def vdetails(self):
        super().vdetails()
        print(f"Number of Doors: {self.doors}")

car1 = Car("Toyota", "Corolla", 2024, 4)

car1.vdetails()


class BankAccount:
    def __init__(self, accountnumber, balance):
        self.accountnumber = accountnumber
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount should be greater than 0.")

     def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance.")
        elif amount <= 0:
            print("Withdrawal amount can not be 0.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. Balance: {self.balance}")

     def display_details(self):
        print(f"Account Number: {self.accountnumber}")
        print(f"Balance: {self.balance}")
        
a1 = BankAccount("ACC1", 5000)
a2 = BankAccount("ACC2", 3000)

a1.deposit(1500)
a2.deposit(2000)

a1.withdraw(2000)
a2.withdraw(1000)

a1.display_details()
a2.display_details()


import math

class Shape:
    def area(self):
        pass   

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

circle = Circle(7)
triangle = Triangle(10, 5)

print("Circle Area:", circle.area())
print("Triangle Area:", triangle.area())


class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary  

    def annualsalary(self):
        return self.salary * 12

class Manager(Employee):
    def __init__(self, name, salary, department, bonus):
        super().__init__(name, salary)
        self.department = department
        self.bonus = bonus  

    def annualsalary(self):
        return (self.salary * 12) + self.bonus


m1 = Manager("Ahsan", 80000, "IT", 200000)
m2 = Manager("Tahir", 90000, "HR", 250000)

print(manager1.name, ":", m1.annual_salary())
print(manager2.name, ":", m2.annual_salary())



class Book:
    def __init__(self, title, author, publication_year):
        self.title = title
        self.author = author
        self.publication_year = publication_year

    def bookdetails(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Publication Year: {self.publication_year}")

class Ebook(Book):
    def __init__(self, title, author, publication_year, price):
        super().__init__(title, author, publication_year)
        self.price = price

    def bookdetails(self):
        super().bookdetails()
        print(f"Price: {self.price}")

ebook1 = Ebook("Python", "Ahsan", 2025, 300)

ebook1.bookdetails()

class Animal:
    def __init__(self, species, sound):
        self.species = species
        self.sound = sound

    def makesound(self):
        print(f"The {self.species} sound: {self.sound}")

class Dog(Animal):
    def __init__(self, species, sound, color):
        super().__init__(species, sound)
        self.color = color

    def makesound(self):
        print(f"The {self.color} {self.species} sound: {self.sound}")


dog1 = Dog("Dog", "Wao", "Black")

dog1.makesound()



class Bank:
    def __init__(self, bankname):
        self.bank_name = bankname
        self.branches = []   

    def addbranch(self, branchname):
        self.branches.append(branchname)
        print(f"Branch '{branchname}' added ")

    def removebranch(self, branchname):
        if branchname in self.branches:
            self.branches.remove(branchname)
            print(f"Branch '{branchname}' removed.")
        else:
            print(f"Branch '{branchname}' not found.")

    def display_branches(self):
        print(f"\nBranches of {self.bankname}:")
        if not self.branches:
            print("No branches found.")
        else:
            for branch in self.branches:
                print(branch)


bank1 = Bank("SMIT Bank")

bank1.addbranch("Islamabad")
bank1.addbranch("Lahore")
bank1.addbranch("Karachi")

bank1.displaybranches()

bank1.removebranch("Lahore")

bank1.displaybranches()


class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price   

    def total_price(self, quantity):
        return self.price * quantity


class PersonalCareProduct(Product):
    def __init__(self, product_id, name, price, warranty_period, warranty_cost):
        super().__init__(product_id, name, price)
        self.warranty_period = warranty_period  
        self.warranty_cost = warranty_cost     

    def total_price(self, quantity):
        base_price = super().total_price(quantity)
        return base_price + self.warranty_cost

product1 = PersonalCareProduct(
    product_id=101,
    name="Electric Machine",
    price=3500,
    warranty_period=2,
    warranty_cost=500
)

quantity = 2
print("Product Name:", product1.name)
print("Warranty Period:", product1.warranty_period, "years")
print("Total Price:", product1.total_price(quantity))



class BankAccount:
    def __init__(self, account_number, holder_name, balance):
        self.account_number = account_number
        self.holder_name = holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited to {self.holder_name}'s account. New balance: {self.balance}")
        else:
            print("Deposit amount should not be 0.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance.")
        elif amount <= 0:
            print("Withdrawal amount should be more than 0")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn from {self.holder_name}'s account. balance: {self.balance}")

    def transfer(self, amount, recipient_account):
        if amount <= 0:
            print("Transfer amount must be greater than 0.")
        elif amount > self.balance:
            print("Insufficient balance for transfer.")
        else:
            self.balance -= amount
            recipient_account.balance += amount
            print(f"{amount} transferred from {self.holder_name} to {recipient_account.holder_name}.")
            print(f"{self.holder_name}'s new balance: {self.balance}")
            print(f"{recipient_account.holder_name}'s new balance: {recipient_account.balance}")

    def display_details(self):
        print(f"Account Number: {self.account_number}")
        print(f"Account Holder: {self.holder_name}")
        print(f"Balance: {self.balance}")
     

account1 = BankAccount("ACC1", "Ahsan", 5000)
account2 = BankAccount("ACC2", "Tahir", 3000)

account1.deposit(2000)
account2.withdraw(1000)
account1.transfer(1500, account2)

account1.display_details()
account2.display_details()


class University:
    def __init__(self, name):
        self.name = name
        self.departments = []  

    def add_department(self, dept_name):
        if dept_name not in self.departments:
            self.departments.append(dept_name)
            print(f"Department '{dept_name}' added ")
        else:
            print(f"Department '{dept_name}' found.")

    def remove_department(self, dept_name):
        if dept_name in self.departments:
            self.departments.remove(dept_name)
            print(f"Department '{dept_name}' removed ")
        else:
            print(f"Department '{dept_name}' not found.")

    def display_departments(self):
        print(f"\nDepartments in {self.name}:")
        if not self.departments:
            print("No departments found.")
        else:
            for dept in self.departments:
                print(dept)

uni = University("Karachi University")

uni.add_department("Computer Science")
uni.add_department("Mathematics")
uni.add_department("Physics")

uni.display_departments()

uni.remove_department("Mathematics")

uni.display_departments()
```
