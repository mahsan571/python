```python
str1=input("Enter any String:")
count=0
count1=0
count2=0
for a in str1:
    if a.isdigit():
        count+=1
    elif a.isalpha():
        count1+=1
    elif a.isprintable():
        count2+=1  
        
print("Total Letters:",count1)    
print("Total Digits:",count)   
print("Total Special Characters:",count2)        
```

```python
str3=input("Enter a Sentence")
count3=0
count4=0
for d in str3:
    if d.isupper():
        count3+=1
    elif d.islower():
        count4+=1

print("Total Uppercase:",count3)
print("Total Lowercase:",count4)    
```

```python
vow1=["a","e","i","o","u"]
vow2=["A","E","I","O","U"]
string10=input("Enter a word:")
for a in string10:
    if a in vow1:
        a=a.replace(a,"*")
    elif a in vow2:
        a=a.replace(a,"*")
    print(a,end="")        
```
